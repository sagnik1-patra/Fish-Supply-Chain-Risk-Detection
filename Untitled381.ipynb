```python
import os
import json
import yaml
import joblib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.callbacks import EarlyStopping

# ============================================================
# PATHS
# ============================================================

BASE_DIR = r"C:\Users\NXTWAVE\Downloads\Fish Supply Chain Risk Detection"
DATA_PATH = os.path.join(BASE_DIR, "ExportoffishandfishproductsinTN_0.csv")
GRAPH_DIR = os.path.join(BASE_DIR, "hybrid_graphs")

os.makedirs(GRAPH_DIR, exist_ok=True)

PREFIX = "hybrid_"

# ============================================================
# LOAD DATA
# ============================================================

df = pd.read_csv(DATA_PATH)
df.columns = df.columns.str.lower()

numeric_cols = df.select_dtypes(include=["int64", "float64"]).columns.tolist()
df = df[numeric_cols].dropna()

# ============================================================
# SCALING
# ============================================================

scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(df)

# ============================================================
# TIME SERIES
# ============================================================

def create_sequences(data, window=5):
    X, y = [], []
    for i in range(len(data) - window):
        X.append(data[i:i + window])
        y.append(np.mean(data[i + window]))
    return np.array(X), np.array(y)

TIME_STEPS = 5
X, y = create_sequences(scaled_data, TIME_STEPS)

split = int(0.8 * len(X))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

# ============================================================
# LSTM MODEL
# ============================================================

model = Sequential([
    LSTM(64, activation="tanh", input_shape=(X.shape[1], X.shape[2])),
    Dense(1)
])

model.compile(optimizer="adam", loss="mse")

history = model.fit(
    X_train, y_train,
    epochs=100,
    batch_size=16,
    validation_split=0.2,
    callbacks=[EarlyStopping(patience=10, restore_best_weights=True)],
    verbose=1
)

# ============================================================
# PREDICTIONS
# ============================================================

test_pred = model.predict(X_test).flatten()

# ============================================================
# CSA  OPTIMIZE AIS THRESHOLD
# ============================================================

def csa_optimize_threshold(predictions, n_nests=20, iterations=30):
    mean = np.mean(predictions)
    std = np.std(predictions)

    nests = np.random.uniform(mean, mean + 3 * std, n_nests)

    def fitness(th):
        risk = np.clip(predictions / th, 0, 1)
        return np.var(risk)   # maximize separation

    for _ in range(iterations):
        new_nests = nests + np.random.normal(0, 0.1, size=n_nests)
        for i in range(n_nests):
            if fitness(new_nests[i]) > fitness(nests[i]):
                nests[i] = new_nests[i]

    best_threshold = nests[np.argmax([fitness(n) for n in nests])]
    return best_threshold

ais_threshold = csa_optimize_threshold(test_pred)

# ============================================================
# AIS RISK SCORING
# ============================================================

risk_scores = np.clip(test_pred / ais_threshold, 0, 1)

# ============================================================
# RESULTS DATAFRAME
# ============================================================

results_df = pd.DataFrame({
    "actual": y_test,
    "predicted": test_pred,
    "risk_score": risk_scores
})

results_df.to_csv(
    os.path.join(BASE_DIR, PREFIX + "risk_results.csv"),
    index=False
)

# ============================================================
# 1 ACCURACY GRAPH
# ============================================================

plt.figure(figsize=(7,5))
plt.plot(history.history["loss"], label="Train Loss")
plt.plot(history.history["val_loss"], label="Validation Loss")
plt.legend()
plt.title("Hybrid AIS + CSA Training Loss")

plt.savefig(os.path.join(GRAPH_DIR, PREFIX + "accuracy_graph.png"), dpi=300, bbox_inches="tight")
plt.show()
plt.close()

# ============================================================
# 2 COMPARISON GRAPH
# ============================================================

plt.figure(figsize=(8,5))
plt.plot(y_test, label="Actual")
plt.plot(test_pred, linestyle="--", label="Predicted")
plt.legend()
plt.title("Hybrid Actual vs Predicted")

plt.savefig(os.path.join(GRAPH_DIR, PREFIX + "comparison_graph.png"), dpi=300, bbox_inches="tight")
plt.show()
plt.close()

# ============================================================
# 3 HEATMAP
# ============================================================

plt.figure(figsize=(6,4))
sns.heatmap(results_df.corr(), annot=True, cmap="coolwarm")
plt.title("Hybrid Correlation Heatmap")

plt.savefig(os.path.join(GRAPH_DIR, PREFIX + "heatmap.png"), dpi=300, bbox_inches="tight")
plt.show()
plt.close()

# ============================================================
# 4 RISK SCORE GRAPH
# ============================================================

plt.figure(figsize=(8,5))
plt.plot(risk_scores, color="red")
plt.axhline(0.8, linestyle="--", label="High Risk")
plt.legend()
plt.title("Hybrid AIS + CSA Risk Scores")

plt.savefig(os.path.join(GRAPH_DIR, PREFIX + "risk_score_graph.png"), dpi=300, bbox_inches="tight")
plt.show()
plt.close()

# ============================================================
# SAVE MODELS
# ============================================================

model.save(os.path.join(BASE_DIR, PREFIX + "lstm_model.h5"))
joblib.dump(scaler, os.path.join(BASE_DIR, PREFIX + "scaler.pkl"))

# ============================================================
# SAVE JSON
# ============================================================

summary = {
    "model": "Hybrid AIS + CSA",
    "ais_threshold": float(ais_threshold),
    "high_risk_routes": int(sum(risk_scores > 0.8)),
    "total_routes": len(risk_scores)
}

with open(os.path.join(BASE_DIR, PREFIX + "prediction_summary.json"), "w") as f:
    json.dump(summary, f, indent=4)

# ============================================================
# SAVE YAML
# ============================================================

yaml_data = {
    "model": "Hybrid AIS + CSA",
    "features_used": numeric_cols,
    "time_steps": TIME_STEPS,
    "threshold": float(ais_threshold)
}

with open(os.path.join(BASE_DIR, PREFIX + "config.yaml"), "w") as f:
    yaml.dump(yaml_data, f)

# ============================================================
# DONE
# ============================================================

print("\n HYBRID AIS + CSA PIPELINE COMPLETED")
print(" Saved in:", BASE_DIR)
print(" Optimized AIS Threshold:", ais_threshold)
```
