```python
import os
import json
import yaml
import joblib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.callbacks import EarlyStopping

# ============================================================
# PATHS & PREFIX
# ============================================================

BASE_DIR = r"C:\Users\NXTWAVE\Downloads\Fish Supply Chain Risk Detection"
DATA_PATH = os.path.join(BASE_DIR, "ExportoffishandfishproductsinTN_0.csv")
GRAPH_DIR = os.path.join(BASE_DIR, "bso_graphs")

PREFIX = "bso_"
os.makedirs(GRAPH_DIR, exist_ok=True)

# ============================================================
# LOAD DATA
# ============================================================

df = pd.read_csv(DATA_PATH)
df.columns = df.columns.str.lower()

numeric_cols = df.select_dtypes(include=["int64", "float64"]).columns.tolist()
df = df[numeric_cols].dropna()

print("âœ” Numeric features used:", numeric_cols)

# ============================================================
# SCALING
# ============================================================

scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(df)

# ============================================================
# TIME SERIES CREATION
# ============================================================

def create_sequences(data, window=5):
    X, y = [], []
    for i in range(len(data) - window):
        X.append(data[i:i + window])
        y.append(np.mean(data[i + window]))
    return np.array(X), np.array(y)

TIME_STEPS = 5
X, y = create_sequences(scaled_data, TIME_STEPS)

split = int(0.8 * len(X))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

# ============================================================
# LSTM MODEL
# ============================================================

model = Sequential([
    LSTM(64, activation="tanh", input_shape=(X.shape[1], X.shape[2])),
    Dense(1)
])

model.compile(optimizer="adam", loss="mse")

history = model.fit(
    X_train, y_train,
    epochs=100,
    batch_size=16,
    validation_split=0.2,
    callbacks=[EarlyStopping(patience=10, restore_best_weights=True)],
    verbose=1
)

# ============================================================
# PREDICTIONS
# ============================================================

test_pred = model.predict(X_test).flatten()

# ============================================================
# BAT ALGORITHM (GLOBAL SEARCH)
# ============================================================

def bat_search(predictions, n_bats=20, iterations=25):
    mean, std = np.mean(predictions), np.std(predictions)
    bats = np.random.uniform(mean, mean + 3 * std, n_bats)
    velocity = np.zeros(n_bats)

    loudness = np.ones(n_bats)
    pulse = np.random.rand(n_bats)

    def fitness(th):
        return np.var(np.clip(predictions / th, 0, 1))

    best = bats[np.argmax([fitness(b) for b in bats])]

    for t in range(iterations):
        for i in range(n_bats):
            freq = np.random.rand()
            velocity[i] += (bats[i] - best) * freq
            bats[i] += velocity[i]

            if np.random.rand() > pulse[i]:
                bats[i] = best + 0.01 * np.random.randn()

            if fitness(bats[i]) > fitness(best) and np.random.rand() < loudness[i]:
                best = bats[i]
                loudness[i] *= 0.9
                pulse[i] *= (1 - np.exp(-0.1 * t))

    return bats

bat_population = bat_search(test_pred)

# ============================================================
# PSO (LOCAL REFINEMENT)
# ============================================================

def pso_refine(predictions, particles, iterations=30):
    velocities = np.random.uniform(-0.1, 0.1, len(particles))

    def fitness(th):
        return np.var(np.clip(predictions / th, 0, 1))

    pbest = particles.copy()
    pbest_scores = np.array([fitness(p) for p in particles])
    gbest = pbest[np.argmax(pbest_scores)]

    w, c1, c2 = 0.7, 1.5, 1.5

    for _ in range(iterations):
        for i in range(len(particles)):
            velocities[i] = (
                w * velocities[i]
                + c1 * np.random.rand() * (pbest[i] - particles[i])
                + c2 * np.random.rand() * (gbest - particles[i])
            )
            particles[i] += velocities[i]

            score = fitness(particles[i])
            if score > pbest_scores[i]:
                pbest[i] = particles[i]
                pbest_scores[i] = score

        gbest = pbest[np.argmax(pbest_scores)]

    return gbest

optimized_threshold = pso_refine(test_pred, bat_population)

# ============================================================
# RISK SCORING
# ============================================================

risk_scores = np.clip(test_pred / optimized_threshold, 0, 1)

# ============================================================
# RESULTS CSV
# ============================================================

results_df = pd.DataFrame({
    "actual": y_test,
    "predicted": test_pred,
    "risk_score": risk_scores
})

results_df.to_csv(
    os.path.join(BASE_DIR, PREFIX + "risk_results.csv"),
    index=False
)

# ============================================================
# 1ï¸âƒ£ ACCURACY GRAPH
# ============================================================

plt.figure(figsize=(7,5))
plt.plot(history.history["loss"], label="Train Loss")
plt.plot(history.history["val_loss"], label="Validation Loss")
plt.legend()
plt.title("Hybrid BA + PSO Training Loss")

plt.savefig(os.path.join(GRAPH_DIR, PREFIX + "accuracy_graph.png"),
            dpi=300, bbox_inches="tight")
plt.show()
plt.close()

# ============================================================
# 2ï¸âƒ£ COMPARISON GRAPH
# ============================================================

plt.figure(figsize=(8,5))
plt.plot(y_test, label="Actual")
plt.plot(test_pred, linestyle="--", label="Predicted")
plt.legend()
plt.title("Hybrid BA + PSO: Actual vs Predicted")

plt.savefig(os.path.join(GRAPH_DIR, PREFIX + "comparison_graph.png"),
            dpi=300, bbox_inches="tight")
plt.show()
plt.close()

# ============================================================
# 3ï¸âƒ£ HEATMAP
# ============================================================

plt.figure(figsize=(6,4))
sns.heatmap(results_df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Hybrid BA + PSO Correlation Heatmap")

plt.savefig(os.path.join(GRAPH_DIR, PREFIX + "heatmap.png"),
            dpi=300, bbox_inches="tight")
plt.show()
plt.close()

# ============================================================
# 4ï¸âƒ£ RISK SCORE GRAPH
# ============================================================

plt.figure(figsize=(8,5))
plt.plot(risk_scores, color="red", label="Risk Score")
plt.axhline(0.8, linestyle="--", color="black", label="High Risk Threshold")
plt.legend()
plt.title("Hybrid BA + PSO Risk Scores")

plt.savefig(os.path.join(GRAPH_DIR, PREFIX + "risk_score_graph.png"),
            dpi=300, bbox_inches="tight")
plt.show()
plt.close()

# ============================================================
# SAVE MODEL FILES
# ============================================================

model.save(os.path.join(BASE_DIR, PREFIX + "lstm_model.h5"))
joblib.dump(scaler, os.path.join(BASE_DIR, PREFIX + "scaler.pkl"))

# ============================================================
# SAVE JSON
# ============================================================

summary = {
    "model": "Hybrid BA + PSO",
    "optimized_threshold": float(optimized_threshold),
    "total_routes": len(risk_scores),
    "high_risk_routes": int(sum(risk_scores > 0.8))
}

with open(os.path.join(BASE_DIR, PREFIX + "prediction_summary.json"), "w") as f:
    json.dump(summary, f, indent=4)

# ============================================================
# SAVE YAML
# ============================================================

yaml_data = {
    "model": "Hybrid BA + PSO",
    "features_used": numeric_cols,
    "time_steps": TIME_STEPS,
    "optimized_threshold": float(optimized_threshold)
}

with open(os.path.join(BASE_DIR, PREFIX + "config.yaml"), "w") as f:
    yaml.dump(yaml_data, f)

# ============================================================
# DONE
# ============================================================

print("\nâœ… HYBRID BA + PSO PIPELINE COMPLETED")
print("ðŸ“ Saved in:", BASE_DIR)
print("âš ï¸ Optimized Threshold:", optimized_threshold)
```
